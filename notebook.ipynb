## Preprocessing

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.columns if data[c].dtype.name != 'object']
```

```python
print(categorical_columns)
```

```python
data = pd.read_csv('young-people-survey/responses.csv', engine='python')
```

###   

```python
data = data.fillna(data.median(axis = 0), axis = 0)
```

```python
data[numerical_columns].count(axis = 0)
```

```python
data[categorical_columns].count(axis = 0)
```

     ,     

```python
data_describe = data.describe(include = [object])
for c in categorical_columns:
    data[c] = data[c].fillna(data_describe[c]['top'])
```

```python
data[categorical_columns].count(axis = 0)
```

```python
data.describe(include = [object])
```

### 

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

```python
data['Gender'].unique()[1]
```

```python
for b in binary_columns:
    data.at[data[b] == data[b].unique()[0], b] = 0
    data.at[data[b] == data[b].unique()[1], b] = 1
    
```

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
data_nonbinary.head()
```

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
```

```python
data_numerical.describe()
```

## C    

```python
data = pd.concat((data_numerical, data_nonbinary, data[binary_columns]), axis = 1)
print(data.shape)
```

```python
data.head()
```

```python
data.describe()
```

### X  y

```python
X = data.drop(('Punctuality_i am always on time'), axis = 1) #   'Punctuality_i am always on time'
y = data['Punctuality_i am always on time']
feature_names = X.columns
```

```python
N, d = X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

# $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

###  

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

```python
best_num_neurons = [(x, y) for x in range(50, 500, 20) for y in range(50, 500, 20)]
test_err = []
train_err = []
n_number = []
for n in best_num_neurons:
    mlp_model = MLPClassifier(hidden_layer_sizes = n, 
                              solver = 'lbfgs', activation = 'logistic', random_state = 13)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    n_number.append(n)

min_test_err = np.min(test_err)
best_num_neurons_arr = [n_number[i] for i in range(0, len(n_number)) if test_err[i] == min_test_err]
print("   = {}".format(best_num_neurons_arr))

alpha_arr = np.linspace(0.0001, 0.1, 150)
test_err = []
train_err = []
for alp in alpha_arr:
    mlp_model = MLPClassifier(alpha = alp, hidden_layer_sizes = best_num_neurons_arr[0],
                              solver = 'lbfgs', activation = 'logistic', random_state = 13)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))

min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)

alpha_opt_arr = alpha_arr[test_err == min_test_err]
print("  alpha = {}".format(alpha_opt_arr))
```

##  

```python
mlp_model = MLPClassifier(alpha = alpha_opt_arr, hidden_layer_sizes = best_num_neurons_arr[0],
                              solver = 'lbfgs', activation = 'logistic', random_state = 13)
mlp_model.fit(X_train, y_train)

err_train = np.mean(y_train != mlp_model.predict(X_train))
err_test  = np.mean(y_test  != mlp_model.predict(X_test))

print(err_train, err_test)
```
