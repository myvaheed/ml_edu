## Preprocessing

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.columns if data[c].dtype.name != 'object']
```

```python
print(categorical_columns)
```

```python
data = pd.read_csv('young-people-survey/responses.csv', engine='python')
```

### Обрабатываем пропущенные значения

Заполняем пропущенные числовые значения количественной переменной их средним значением 

```python
data = data.fillna(data.median(axis = 0), axis = 0)
```

```python
data[numerical_columns].count(axis = 0)
```

```python
data[categorical_columns].count(axis = 0)
```

Заполняем пропущенные значения категориальной переменной значением, которое наиболее часто встречается 

```python
data_describe = data.describe(include = [object])
for c in categorical_columns:
    data[c] = data[c].fillna(data_describe[c]['top'])
```

```python
data[categorical_columns].count(axis = 0)
```

```python
data.describe(include = [object])
```

### Векторизация

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

```python
data['Gender'].unique()[1]
```

```python
for b in binary_columns:
    data.at[data[b] == data[b].unique()[0], b] = 0
    data.at[data[b] == data[b].unique()[1], b] = 1
    
```

Для небинарных категориальных признаков применяем метод векторизации

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
data_nonbinary.head()
```

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
```

```python
data_numerical.describe()
```

## Cоединяем все в одну таблицу

```python
data = pd.concat((data_numerical, data_nonbinary, data[binary_columns]), axis = 1)
print(data.shape)
```

```python
data.head()
```

```python
data.describe()
```

### X и y

```python
X = data.drop(('Punctuality_i am always on time'), axis = 1) # выбрасываем столбец 'Punctuality_i am always on time'
y = data['Punctuality_i am always on time']
feature_names = X.columns
```

```python
N, d = X.shape
```

Разбиваем данные на тестовую и обучающую выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

# $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

### Подбор параметров

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

## Еще раз проверим значимость признаков

```python
#функция графики значимости признаков по порядку убывания
def plot_importances(classifier, feature_names, top_features=20):
 coef = classifier.coef_.ravel()
 top_positive_coefficients = np.argsort(coef)[-top_features:]
 top_negative_coefficients = np.argsort(coef)[:top_features]
 top_coefficients = np.hstack([top_negative_coefficients, top_positive_coefficients])
 # create plot
 plt.figure(figsize=(15, 5))
 colors = ['red' if c < 0 else 'blue' for c in coef[top_coefficients]]
 plt.bar(np.arange(2 * top_features), coef[top_coefficients], color=colors)
 feature_names = np.array(feature_names)
 plt.xticks(np.arange(1, 1 + 2 * top_features), feature_names[top_coefficients], rotation=60, ha='right')
 plt.show()  
```

```python
def f_importances(coef, names):
    imp = coef
    zimp_names = zip(imp,names)
    srtd = sorted(zimp_names,key=lambda x:x[1])
    imp,names = zip(*srtd)
    plt.barh(range(len(names)), imp, align='center')
    plt.yticks(range(len(names)), names)
    plt.show()
```
